```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("recovery-news-data.csv")

# Convert reliability to 3-tier trust score
def get_trust_tier(row):
    if row['news_guard_score'] >= 75 and row['mbfc_level'] == 'High':
        return 2  # High trust
    elif row['news_guard_score'] >= 50 or row['mbfc_level'] == 'Mixed':
        return 1  # Medium trust
    else:
        return 0  # Low trust

df['trust_tier'] = df.apply(get_trust_tier, axis=1)

# Split data
train_df, val_df = train_test_split(df, test_size=0.2, stratify=df['trust_tier'])
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import torch
from sklearn.metrics import accuracy_score

class TrustDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

class TrustworthinessModel:
    def __init__(self):
        self.tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
        self.model = BertForSequenceClassification.from_pretrained(
            'bert-base-uncased', num_labels=3)
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.model.to(self.device)

    def train(self, train_texts, train_labels, val_texts, val_labels):
        # Tokenize texts
        train_encodings = self.tokenizer(train_texts, truncation=True, padding=True, max_length=512)
        val_encodings = self.tokenizer(val_texts, truncation=True, padding=True, max_length=512)

        # Create datasets
        train_dataset = TrustDataset(train_encodings, train_labels)
        val_dataset = TrustDataset(val_encodings, val_labels)

        # Training arguments
        training_args = TrainingArguments(
            output_dir='./results',
            num_train_epochs=3,
            per_device_train_batch_size=8,
            per_device_eval_batch_size=16,
            warmup_steps=500,
            weight_decay=0.01,
            evaluation_strategy='epoch',
            logging_dir='./logs',
        )

        # Trainer
        trainer = Trainer(
            model=self.model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
            compute_metrics=lambda p: {'accuracy': accuracy_score(p.label_ids, p.predictions.argmax(-1))}
        )

        trainer.train()
        self.model.save_pretrained('./trust_model')
        self.tokenizer.save_pretrained('./trust_model')

    def predict(self, text):
        inputs = self.tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
        inputs = {k: v.to(self.device) for k, v in inputs.items()}
        with torch.no_grad():
            outputs = self.model(**inputs)
        return torch.argmax(outputs.logits, dim=1).cpu().item()


# Initialize and train
trust_model = TrustworthinessModel()
trust_model.train(
    train_df['body_text'].tolist(),
    train_df['trust_tier'].tolist(),
    val_df['body_text'].tolist(),
    val_df['trust_tier'].tolist()
)

```

```python
import gradio as gr
import pandas as pd
import torch
import faiss
import numpy as np
from transformers import (
    BertTokenizer, BertForSequenceClassification,
    DPRQuestionEncoder, DPRContextEncoder,
    BartForConditionalGeneration, BartTokenizer,
    DPRQuestionEncoderTokenizer, DPRContextEncoderTokenizer
)

class TrustRAGChatbot:
    def __init__(self, data_path="recovery-news-data.csv"):
        self.documents = self._load_and_process_data(data_path)
        self.trust_model = self._load_trust_model()
        self.retriever = TrustAwareRetriever(self.trust_model)
        self.generator = TrustAwareGenerator()
        self.retriever.build_index(self.documents)

    def _load_and_process_data(self, data_path):
        df = pd.read_csv(data_path)

        # Convert reliability scores to trust tiers
        def calculate_trust_tier(row):
            if row['news_guard_score'] >= 75 and row['mbfc_level'] in ['High', 'Very High']:
                return 2  # High trust
            elif row['news_guard_score'] >= 50 or row['mbfc_level'] in ['Mixed', 'Medium']:
                return 1  # Medium trust
            else:
                return 0  # Low trust

        df['trust_tier'] = df.apply(calculate_trust_tier, axis=1)

        return [
            {
                "content": row['body_text'],
                "metadata": {
                    'source': row['publisher'],
                    'url': row['url'],
                    'date': row['publish_date'],
                    'author': row['author'],
                    'political_bias': row['political_bias'],
                    'country': row['country'],
                    'trust_tier': row['trust_tier'],
                    'news_guard_score': row['news_guard_score'],
                    'mbfc_level': row['mbfc_level']
                }
            }
            for _, row in df.iterrows()
        ]

    def _load_trust_model(self):
        model = BertForSequenceClassification.from_pretrained('./trust_model')
        tokenizer = BertTokenizer.from_pretrained('./trust_model')
        return TrustworthinessModel(model, tokenizer)

    def chat(self, query, history):
        retrieved_docs = self.retriever.retrieve(query)
        response = self.generator.generate(query, retrieved_docs)

        # Build detailed source information
        sources = []
        for doc, score in retrieved_docs[:3]:
            source_info = f"""• {doc['metadata']['source']}
            - Trust Level: {self._get_trust_label(doc['metadata']['trust_tier'])}
            - NewsGuard Score: {doc['metadata']['news_guard_score']}
            - MBFC Rating: {doc['metadata']['mbfc_level']}
            - Political Bias: {doc['metadata']['political_bias']}
            - Country: {doc['metadata']['country']}
            - URL: {doc['metadata']['url']}"""
            sources.append(source_info)

        warning = self._generate_warnings(retrieved_docs)

        full_response = f"{response}\n\n=== Source Reliability Information ===\n" + "\n\n".join(sources)
        if warning:
            full_response = f"⚠️ {warning}\n\n" + full_response

        return full_response

    def _get_trust_label(self, tier):
        return ["Low", "Medium", "High"][tier]

    def _generate_warnings(self, retrieved_docs):
        warnings = []
        low_trust = sum(1 for doc, _ in retrieved_docs if doc['metadata']['trust_tier'] == 0)
        mixed_bias = len(set(doc['metadata']['political_bias'] for doc, _ in retrieved_docs)) > 1

        if low_trust > 0:
            warnings.append(f"Contains {low_trust} low-reliability sources")
        if mixed_bias:
            warnings.append("Sources show political bias diversity")

        return " | ".join(warnings) if warnings else ""

class TrustAwareRetriever:
    def __init__(self, trust_model, top_k=5):
        self.question_encoder = DPRQuestionEncoder.from_pretrained("facebook/dpr-question_encoder-single-nq-base")
        self.question_tokenizer = DPRQuestionEncoderTokenizer.from_pretrained("facebook/dpr-question_encoder-single-nq-base")
        self.context_encoder = DPRContextEncoder.from_pretrained("facebook/dpr-ctx_encoder-single-nq-base")
        self.context_tokenizer = DPRContextEncoderTokenizer.from_pretrained("facebook/dpr-ctx_encoder-single-nq-base")
        self.trust_model = trust_model
        self.top_k = top_k
        self.index = None
        self.doc_store = []

    def build_index(self, documents):
        embeddings = []
        self.doc_store = documents

        for doc in documents:
            trust_weight = doc["metadata"]["trust_tier"] / 2
            enhanced_content = f"[NewsGuard: {doc['metadata']['news_guard_score']}] [MBFC: {doc['metadata']['mbfc_level']}] {doc['content']}"

            # Tokenize and encode
            inputs = self.context_tokenizer(
                enhanced_content,
                return_tensors="pt",
                max_length=256,
                truncation=True
            )
            embedding = self.context_encoder(**inputs).pooler_output[0].detach().numpy()

            weighted_embedding = embedding * (1 + trust_weight)
            embeddings.append(weighted_embedding)

        embeddings = np.array(embeddings).astype('float32')
        self.index = faiss.IndexFlatIP(embeddings.shape[1])
        self.index.add(embeddings)

    def retrieve(self, query):
        # Tokenize and encode query
        inputs = self.question_tokenizer(
            query,
            return_tensors="pt",
            max_length=256,
            truncation=True
        )
        query_embedding = self.question_encoder(**inputs).pooler_output[0].detach().numpy().astype('float32')

        scores, indices = self.index.search(np.expand_dims(query_embedding, 0), self.top_k*3)

        results = []
        for score, idx in zip(scores[0], indices[0]):
            doc = self.doc_store[idx]
            trust_score = doc["metadata"]["trust_tier"] / 2
            combined_score = (0.5 * score) + (0.3 * trust_score) + (0.2 * (doc["metadata"]["news_guard_score"]/100))
            results.append((doc, combined_score))

        filtered = [r for r in results if r[0]["metadata"]["trust_tier"] > 0] or results
        return sorted(filtered, key=lambda x: x[1], reverse=True)[:self.top_k]

class TrustAwareGenerator:
    def __init__(self):
        self.model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
        self.tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.model.to(self.device)

    def generate(self, query, retrieved_docs):
        context = []
        for doc, score in retrieved_docs:
            if doc["metadata"]["trust_tier"] >= 1:
                context.append({
                    "text": doc["content"][:1000],
                    "reliability": f"NewsGuard: {doc['metadata']['news_guard_score']}, MBFC: {doc['metadata']['mbfc_level']}"
                })

        if not context:
            return "No sufficiently reliable sources found to answer this question."

        # Create structured context
        context_str = "\n".join(
            f"[Source {i+1} | Reliability: {c['reliability']}]\n{c['text']}"
            for i, c in enumerate(context[:3])
        )
        input_text = f"Question: {query}\n\nReliable Context:\n{context_str}"

        inputs = self.tokenizer(input_text, return_tensors="pt", max_length=1024, truncation=True)
        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        outputs = self.model.generate(
            inputs["input_ids"],
            max_length=400,
            num_beams=4,
            early_stopping=True,
            no_repeat_ngram_size=3
        )

        return self.tokenizer.decode(outputs[0], skip_special_tokens=True)

```

```python
# Create and launch interface
chatbot = TrustRAGChatbot()

def respond(message, history):
    return chatbot.chat(message, history)

interface = gr.ChatInterface(
    respond,
    title="Trust-Verified Information Assistant",
    description="Ask any question and receive answers with verified reliability information",
    examples=[
        ["What are the health effects of vaping?"],
        ["Tell me about recent climate change policies"],
        ["Is there evidence for vaccine side effects?"]
    ],
    css=".warning { color: red; }"
)

if __name__ == "__main__":
    interface.launch()
```
